## 3)  Exploratory Data Analysis and Visualization

```r
players<- read_csv("players.csv")
```

```r
tidy_players<- players |>
mutate(subscribe = as_factor(subscribe),
      experience = as_factor(experience))|>
filter(played_hours>0.0)|>
filter(!is.na(Age))
tidy_players
```

The players.csv data is already quite tidy with one variable in every column, one observation in every row, and one value per cell. Therefore, I have just removed the NA/invalid values and turn experience and subscribe into factors because they are categorical values and this would make it easier to create graphs.

```r
players|>
summarise(age_mean = round(mean(Age,na.rm=TRUE),2),
         played_hours_mean = round(mean(played_hours,na.rm=TRUE), 2))
```

```r
tidy_players|>
summarise(age_mean = round(mean(Age,na.rm=TRUE),2),
         played_hours_mean = round(mean(played_hours,na.rm=TRUE), 2))
```

```r
plot_1<- tidy_players|>
ggplot(aes(x=Age, y=played_hours, colour=subscribe))+
geom_point(size=3, alpha=0.6)+
labs(x="Age (years)", y="Hours Playing Video Game", colour = "Subcribed or Not", title="Age vs. Hours Played by Newsletter Subscription")
plot_1
```

### Plot 1
This plot shows the relationship between hours played and age of the player, coloured by whether they subscribe to gaming newsletters. Through this graph, we can tell that there is a slight preference of those who play less to not subscribe to newsletters. However, there is still a large overlap between the people who have subscribed, which signifies a weak relationship between the variables. In addition, looking at just the axis, we can tell that playing video games is more popular among the ages 15-25.

```r
plot_2<- tidy_players|>
ggplot(aes(x=experience, fill=subscribe))+
geom_bar(position= "dodge")+
labs(x="Experience in Video Games", y="Number of People", fill = "Subcribed or Not", title="Newsletter Subscription by Experience")
plot_2
```

### Plot 2 

This plot shows the number of newsletter subscribers at different skill levels. This is relevant to my question, as it shows that the proportion of subscribers changes based on experience, not just age and hours played. Some experience levels (e.g., Amateur) have more subscribers solely due to more people in that group, not necessarily because there is a higher proportion of subscribers. A 100% stacked bar chart would be better to use for comparing proportions of subscribers across experience levels. Therefore, this plot indicates that experience level should be used in later analysis, as the relationship between hours played and subscription may not be the same for every group.

```r
plot_3<- tidy_players|>
ggplot(aes(x=experience, fill=subscribe))+
geom_bar(position="fill")+
labs(x="Experience in Video Games", y="Number of People", fill = "Subcribed or Not", title="Newsletter Subscription by Experience")
plot_3
```

### Plot 3

This plot shows the proportion of players who subscribed to the newsletter within their skill level. This plot builds on plot 2 as it expands off a different aspect of the question: will the number of subscriptions change depending on skill level? In this graph, we can tell that regular players have the highest proportion of subscribers. This is different from my expectation of the data; I had expected that the higher-skilled people (Pro, Veteran) would subscribe more, but the data suggests the opposite, they are less likely out of the groups to subscribe to gaming newsletters.
